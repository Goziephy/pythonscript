```python
import matplotlib.pyplot as plt
import numpy as np
# Calculation of Allowable Bearing Presure (Qa) from Vp and Vs Dataset
Vs = np.loadtxt("Linder_S_finamod.txt")
Vp = np.loadtxt("Linder_P_finamod.txt")
gam_0 = 16
n = 4
gam = gam_0 + 0.002*Vp
Qa = 0.1*gam*Vs/n
np.savetxt("Linder_Qa_finamod.txt", Qa, fmt='%.18e', delimiter=' ', newline='\n', header='', footer='')
# figure 1: Visualization of P-Wave Velocity Dataset
Vp = np.loadtxt("Linder_P_finamod.txt")
plt.imshow(Vp,origin='lower',extent =[0,30,250,0],aspect='auto',cmap='jet')
plt.colorbar()
plt.xlabel("Distance (m)")
plt.ylabel("Depth (m)")
plt.title("P-Wave Velocity Model (Vp)")
plt.show()
# figure 2: Visualization of S-Wave Velocity Dataset
Vs = np.loadtxt("Linder_S_finamod.txt")
plt.imshow(Vs,origin='lower',extent =[0,30,250,0],aspect='auto',cmap='jet')
plt.colorbar()
plt.xlabel("Distance (m)")
plt.ylabel("Depth (m)")
plt.title("S-Wave Velocity Model (Vs)")
plt.show()
# figure 3: Visualization of Allowable Bearing Pressure Dataset
Qa = np.loadtxt("Linder_Qa_finamod.txt")
plt.imshow(Qa,origin='lower',extent =[0,30,250,0],aspect='auto',cmap='jet')
plt.colorbar()
plt.xlabel("Distance (m)")
plt.ylabel("Depth (m)")
plt.title("Allowable Bearing Pressure Model (Qa)")
plt.show()

print('Shape of Vp,Vs,Qa = ',Vp.shape,Vs.shape,Qa.shape)
```
